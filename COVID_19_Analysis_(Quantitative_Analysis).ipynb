```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
df = pd.read_csv("https://raw.githubusercontent.com/WidhyaOrg/datasets/master/covid19.csv")
df.tail()
```

```
### sorting the data accordint to date
df1= df.groupby('Date', sort = False).sum()
df1
```

```
df1['total_cases'] = df1['ConfirmedIndianNational'] + df1['ConfirmedForeignNational'] + df1['Cured'] + df1['Deaths']
df1
```

```
### total no. of cases on 4 march 2020
cases = int(df1.groupby('Date').get_group('04/03/20')['total_cases'])
cases
```

```
### Visualizing the data using matplotlib
x = df1.reset_index()['Date']
y = df1['total_cases']
```

```
plt.figure(figsize = (15,5))
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.xticks(rotation = 90)
plt.title("TOTAL CASES DATE WISE")
plt.plot(x,y)
```

```
x = df1.reset_index()['Date']
y = df1['Deaths']
plt.figure(figsize = (15,5))
plt.xticks(rotation = 90)
plt.title("Deaths date wise")
plt.xlabel("Date")
plt.ylabel('No. of Deaths')
plt.plot(x,y)
```
